```python
from itertools import combinations

# 복도의 크기
n = int(input())

# 복도의 정보 (n x n)
board = []

# 모든 선생님 위치 정보
teachers = []

# 모든 빈 공간의 위치 정보
spaces = []

for i in range(n):
    board.append(list(input().split()))
    
    for j in range(n):
        # 선생님이 존재하는 위치 저장
        if board[i][j] == 'T':
            teachers.append((i,j))
        
        # 장애물을 설치할 수 있는 (빈 공간) 위치 저장
        if board[i][j] == 'X':
            spaces.append((i,j))
            
# 특정 방향으로 감시를 진행하여, 학생이 발견될 경우 True, 학생이 발견되지 않을 경우 False
def watch(x, y, direction):
    
    # 왼쪽 방향으로 감시
    if direction == 0:
        # 현재의 열의 인덱스부터 열의 인덱스가 0보다 클 때까지 탐색한다면 왼쪽 모두를 탐색할 수 있음
        while y >= 0:
            # 학생인 경우
            if board[x][y] == 'S':
                return True
            # 장애물이 있는 경우
            if board[x][y] == 'O':
                return False
            
            y -= 1
            
    # 오른쪽 방향으로 감시
    if direction == 1:
        # 현재의 열의 인덱스부터 열의 인덱스가 n보다 작을 때까지 탐색한다면 오른쪽 모두를 탐색할 수 있음
        while y < n:
            # 학생인 경우
            if board[x][y] == 'S':
                return True
            # 장애물이 있는 경우
            if board[x][y] == 'O':
                return False
            
            y += 1
            
    # 위쪽 방향으로 감시
    if direction == 2:
        # 현재의 행의 인덱스부터 행의 인덱스가 0보다 클 때까지 탐색한다면 위쪽 모두를 탐색할 수 있음
        while x >= 0:
            # 학생인 경우
            if board[x][y] == 'S':
                return True
            # 장애물이 있는 경우
            if board[x][y] == 'O':
                return False
            
            x -= 1
    
    # 아래쪽 방향으로 감시
    if direction == 3:
        # 현재의 행의 인덱스부터 행의 인덱스가 n보다 작을 때까지 탐색한다면 아래쪽 모두를 탐색할 수 있음
        while x < n:
            # 학생인 경우
            if board[x][y] == 'S':
                return True
            # 장애물이 있는 경우
            if board[x][y] == 'O':
                return False
            
            x += 1
            
    return False

# 장애물 설치 이후에, 한 명이라도 학생이 감지되는지 검사하는 함수
def process():
    
    # 모든 선생님의 위치를 하나씩 확인
    for x, y in teachers:
        
        # 4가지 방향으로 학생을 감지할 수 있는지 확인
        for i in range(4):
            if watch(x, y, i):
                return True
            
    return False
    
# 학생이 한 명도 감지되지 않도록 설치할 수 있는지의 여부
find = False
    
# 빈 공간에서 3개를 뽑는 모든 조합을 확인
for data in combinations(spaces, 3):
    
    # 장애물 설치해보기
    for x, y in data:
        board[x][y] == 'O'
    
    # 학생이 한 명도 감지되지 않는 경우
    if not process():
        # 원하는 경우를 발견한 것
        find = True
        break
            
    # 설치된 장애물을 다시 없애기
    for x, y in data:
        board[x][y] = 'X'

if find:
    print('YES')
else:
    print('NO')
            
```
